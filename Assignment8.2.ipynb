Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

Ans. Seaborn is a popular Python data visualization library built on top of Matplotlib. It provides a high-level interface for creating informative and attractive statistical graphics. Here are five common types of plots you can create using Seaborn along with their typical use cases:

Scatter Plot:
These are used to visualize the relationship between two continuous variables. They help you understand patterns, correlations, and outliers in your data.

Line Plot:
These are used to show how a single continuous variable changes over time or across different categories. They are useful for visualizing trends and patterns.

Bar Plot:
These are used to display the distribution of categorical data. They show the frequency or count of each category and are useful for comparing different categories.

Histogram:
These are used to visualize the distribution of a single continuous variable. They help you understand the underlying data distribution, including its shape and central tendencies.

Box Plot (Box-and-Whisker Plot):
These are used to display the distribution and spread of a continuous variable, showing key statistics such as the median, quartiles, and potential outliers. They are useful for identifying variability and skewness in the data.

Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x =
"timepoint" and y = "signal" for different events and regions.
Note: timepoint, signal, event, and region are columns in the fmri dataset.

```python
sns.load_dataset("fmri")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

fmri_data = sns.load_dataset("fmri")

plt.figure(figsize=((5,3)))  

sns.lineplot(data=fmri_data, x="timepoint", y="signal", hue="event", style="region")

plt.xlabel("Timepoint")
plt.ylabel("Signal")
plt.title("Line Plot of Signal Over Time by Event and Region")

plt.legend(title="Event", loc="upper right")
plt.show()
```

Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x =
'pclass', y = 'age' and y = 'fare'.
Note: pclass, age, and fare are columns in the titanic dataset.

```python
sns.load_dataset("titanic")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

titanic_data = sns.load_dataset("titanic")

plt.figure(figsize=(8, 7))

plt.subplot(1, 2, 1)
sns.boxplot(data=titanic_data, x='pclass', y='age')
plt.xlabel('Pclass')
plt.ylabel('Age')
plt.title('Box Plot of Age by Pclass')


plt.subplot(1, 2, 2)
sns.boxplot(data=titanic_data, x='pclass', y='fare')
plt.xlabel('Pclass')
plt.ylabel('Fare')
plt.title('Box Plot of Fare by Pclass')

plt.tight_layout()
plt.show()
```

Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue parameter for the 'cut' column of the diamonds dataset.

```python
sns.load_dataset("diamonds")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt


diamonds_data = sns.load_dataset("diamonds")

plt.figure(figsize=(7, 5)) 

sns.histplot(data=diamonds_data, x='price', hue='cut', bins=30, kde=True)

plt.xlabel('Price')
plt.ylabel('Frequency')
plt.title('Histogram of Price with Hue for Cut')


plt.legend(title='Cut', loc='upper right')
plt.show()
```

Que 6: Use the "flights" dataset from seaborn to plot a heatmap.

```python
sns.load_dataset("flights")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

flights_data = sns.load_dataset("flights")

flights_pivot = flights_data.pivot(index="month", columns="year", values="passengers")

plt.figure(figsize=(10, 8)) 
sns.heatmap(data=flights_pivot, annot=True, fmt="d", cmap="plasma", linewidths=.5)

plt.xlabel("Year")
plt.ylabel("Month")
plt.title("Flight Passenger Counts Over Time (1950-1960)")

plt.show()
```
