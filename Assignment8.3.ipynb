Q1. Load the "titanic" dataset using the load_dataset function of seaborn. Use Plotly express to plot a
scatter plot for age and fare columns in the titanic dataset.

```python
import seaborn as sns
import plotly.express as px

titanic_data = sns.load_dataset("titanic")

fig = px.scatter(titanic_data, x="age", y="fare", title="Age vs. Fare")

fig.update_layout(width=800, height=500)
fig.show()
```

Q2. Using the tips dataset in the Plotly library, plot a box plot using Plotly express.

```python
import seaborn as sns
import plotly.express as px
tips_data = sns.load_dataset('tips')
fig = px.box(tips_data, x='smoker', y='tip', title = 'Smoker vs Tip')
fig.update_layout(width=700, height=700)
fig.show()
```

Q3. Using the tips dataset in the Plotly library, Plot a histogram for x= "sex" and y="total_bill" column in the tips dataset. Also, use the "smoker" column with the pattern_shape parameter and the "day" column with the color parameter.

```python
import seaborn as sns
import plotly.express as px
tips_data = sns.load_dataset('tips')
fig = px.histogram(tips_data, x= "sex", y="total_bill", pattern_shape = "smoker", color = "day")

fig.update_layout(width=800, height=600)

fig.show()
```

Q4. Using the iris dataset in the Plotly library, Plot a scatter matrix plot, using the "species" column for the color parameter.
Note: Use "sepal_length", "sepal_width", "petal_length", "petal_width" columns only with the
dimensions parameter.

```python
import plotly.express as px
import seaborn as sns
import warnings
iris_data = sns.load_dataset('iris')

# Suppress the FutureWarning
warnings.filterwarnings('ignore', category=FutureWarning) 

fig = px.scatter_matrix(
    iris_data,
    dimensions=['sepal_length', 'sepal_width', 'petal_length', 'petal_width'],
    color='species',
    title='Scatter Matrix Plot of Iris Dataset',
    labels={'species': 'Species'},
)

fig.update_layout(width=800, height=600)

fig.show()
```

Q5. What is Distplot? Using Plotly express, plot a distplot.

Ans. A distplot, short for "distribution plot," is a graphical representation used to visualize the distribution of a univariate dataset. It displays a histogram with a kernel density estimate (KDE) overlaid on top to provide a smooth approximation of the probability density function of the data. Distplots are useful for understanding the underlying distribution of a dataset, including its central tendency and spread.

```python
import plotly.express as px
import numpy as np

data = np.random.randn(1000)

fig = px.histogram(data, nbins=30, title="Distplot of Sample Data")

fig.update_layout(width=800, height=500)
fig.show()
```
