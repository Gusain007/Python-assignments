Q1. What is the Probability density function?

Ans. The Probability Density Function (PDF) is a fundamental concept in probability and statistics. It describes the likelihood of a continuous random variable taking on a specific value or falling within a range of values. In other words, the PDF provides a way to model the probability distribution of a continuous random variable.

Key characteristics of a probability density function:

(i) Non-negative Values: The PDF must be non-negative for all possible values of the random variable: f(x) ≥ 0 for all x.

(ii) Area under the Curve: The total area under the PDF curve over the entire range of possible values equals 1. This means that the integral of the PDF over its entire domain is 1: ∫f(x) dx = 1.

(iii) Continuous Random Variable: The PDF is used for continuous random variables, as opposed to the Probability Mass Function (PMF), which is used for discrete random variables.

The PDF is typically represented as a continuous curve, and it provides a way to understand the likelihood of different outcomes for a continuous random variable. You can use the PDF to calculate probabilities associated with specific events or ranges of values by integrating the PDF over those ranges.

Q2. What are the types of Probability distribution?

Ans. Probability distributions are mathematical functions that describe the likelihood of different outcomes in a random process. There are several types of probability distributions, and they are broadly categorized into two main types: discrete probability distributions and continuous probability distributions.

Discrete Probability Distributions:

1. Bernoulli Distribution: This distribution models a random experiment with two possible outcomes, typically denoted as "success" and "failure." It is used for situations with binary outcomes, such as coin flips or success/failure trials.

2. Binomial Distribution: The binomial distribution is used when you have a sequence of independent Bernoulli trials. It describes the number of successes (or failures) in a fixed number of trials.

3. Poisson Distribution: The Poisson distribution models the number of events occurring in a fixed interval of time or space when these events happen with a known average rate but independently. It's often used in situations like counting the number of phone calls at a call center in an hour.

4. Geometric Distribution: The geometric distribution models the number of Bernoulli trials needed for the first success to occur. It's used in situations like finding the number of attempts to achieve success in a sequence of trials.

5. Hypergeometric Distribution: This distribution is used for sampling without replacement. It describes the probability of drawing a specific number of "successes" from a finite population without replacement.


Continuous Probability Distributions:

1. Normal Distribution (Gaussian Distribution): The normal distribution is one of the most common continuous probability distributions. It is symmetric and bell-shaped, and it describes a wide range of natural phenomena, making it useful in various fields, including statistics and science.

2. Uniform Distribution: In a uniform distribution, all values within a certain range are equally likely. For example, the continuous uniform distribution describes the probability of a continuous random variable taking on any value within a specified interval.

3. Exponential Distribution: The exponential distribution models the time between events in a Poisson process. It is often used to describe the waiting times between rare events occurring at a constant rate.

4. Log-Normal Distribution: The log-normal distribution is used when the logarithm of a variable follows a normal distribution. It is commonly used to model financial data, among other applications.

5. Gamma Distribution: The gamma distribution is a versatile distribution used for modeling various processes, such as the waiting times until a Poisson number of events occur.

6. Chi-Squared Distribution: The chi-squared distribution is used in hypothesis testing, confidence intervals, and estimation of population variance. It arises frequently in statistical inference.

Q3. Write a Python function to calculate the probability density function of a normal distribution with given mean and standard deviation at a given point.

```python
from scipy.stats import norm
import matplotlib.pyplot as plt
import numpy as np

def normal_pdf(x, mean, std_dev):
    pdf = norm.pdf(x, loc=mean, scale=std_dev)
    return pdf


mean = 2.0
std_dev = 0.5
x = 2.5
pdf_value = normal_pdf(x, mean, std_dev)

x_values = np.linspace(0, 4, 400)
pdf_curve = norm.pdf(x_values, loc=mean, scale=std_dev)

plt.plot(x_values, pdf_curve, label=f'N({mean}, {std_dev}) PDF')
plt.scatter(x, pdf_value, color='red', label=f'PDF at x = {x}: {pdf_value:.4f}')
plt.xlabel('x')
plt.ylabel('PDF')
plt.legend()
plt.title('Normal Distribution PDF')
plt.grid(True)
plt.show()
```

Q4. What are the properties of Binomial distribution? Give two examples of events where binomial distribution can be applied.

Ans. The binomial distribution is a discrete probability distribution that describes the number of successes in a fixed number of independent Bernoulli trials (experiments with two possible outcomes, often referred to as "success" and "failure"). The binomial distribution has several key properties:

a) Discreteness: The binomial distribution deals with discrete random variables, meaning that it counts the number of successes in a fixed number of trials.

b) Two Possible Outcomes: Each trial has two possible outcomes, typically labeled as "success" and "failure."

c) Independence: The trials are assumed to be independent, meaning the outcome of one trial does not affect the outcome of another.

d) Fixed Number of Trials (n): The number of trials (denoted as 'n') is fixed and known in advance.

e) Constant Probability of Success (p): The probability of success (denoted as 'p') is constant for each trial.

f) Discrete Probability Mass Function: The probability mass function (PMF) of the binomial distribution is given by the formula: P(X = k) = C(n, k) * p^k * (1 - p)^(n - k), where X is the random variable representing the number of successes, k is the specific number of successes, n is the total number of trials, p is the probability of success, and C(n, k) represents the binomial coefficient.

Two examples where the binomial distribution can be applied:

a) Coin Flips: When flipping a fair coin (with a probability of success, i.e., getting heads, p = 0.5) multiple times, you can use the binomial distribution to calculate the probability of getting a specific number of heads in a fixed number of flips (e.g., the probability of getting exactly 3 heads in 5 coin flips).

b) Medical Diagnosis: In medical testing, the binomial distribution can be used to model the probability of detecting a specific number of patients with a certain condition among a fixed number of patients who are tested, assuming the tests are independent and the probability of a positive test result is known.

Q5. Generate a random sample of size 1000 from a binomial distribution with probability of success 0.4 and plot a histogram of the results using matplotlib.

Q6. Write a Python function to calculate the cumulative distribution function of a Poisson distribution with given mean at a given point.

```python
from scipy.stats import poisson

def poisson_cdf(k, mean):
    cdf = poisson.cdf(k, mu=mean)
    return cdf

mean = 3.5
k = 2
cdf_value = poisson_cdf(k, mean)
print(f"CDF at k = {k}: {cdf_value:.4f}")
```

Q7. How Binomial distribution different from Poisson distribution?

Ans. The Binomial distribution and the Poisson distribution are both used to model the distribution of discrete random variables, but they have different characteristics and are applied in different scenarios. Here are the key differences between the two distributions:

1. Nature of Random Variable:

Binomial Distribution: The Binomial distribution models the number of successes in a fixed number of independent Bernoulli trials (experiments with two possible outcomes, often referred to as "success" and "failure"). It deals with a fixed number of trials (n).

Poisson Distribution: The Poisson distribution models the number of events occurring in a fixed interval of time or space. It is used when events happen randomly, independently, and at a constant average rate but are rare or infrequent. The number of events is not fixed; it can be any non-negative integer.

2. Parameterization:

Binomial Distribution: It is parameterized by two values, 'n' (the number of trials) and 'p' (the probability of success in each trial).

Poisson Distribution: It is parameterized by a single value, 'λ' (lambda), which represents the average rate of event occurrences within the fixed interval.

3. Probability Mass Function:

Binomial Distribution: The PMF of the Binomial distribution is given by the binomial coefficient and is expressed as P(X = k) = C(n, k) * p^k * (1 - p)^(n - k), where X is the random variable representing the number of successes, and k is the specific number of successes.

Poisson Distribution: The PMF of the Poisson distribution is given by P(X = k) = (e^(-λ) * λ^k) / k!, where X is the random variable representing the number of events, and k is the specific number of events.

4. Use Cases:

Binomial Distribution: The Binomial distribution is appropriate when you have a fixed number of trials (n), each with two possible outcomes, and you want to model the number of successes in those trials. It is commonly used in scenarios like coin flips, success/failure experiments, or when calculating the probability of a certain number of successes in a specific number of trials.

Poisson Distribution: The Poisson distribution is used when you want to model the number of rare and randomly occurring events in a fixed interval. It's suitable for situations where events are infrequent and the average rate of occurrence is known, such as modeling the number of customer arrivals at a store in a given time period or the number of email arrivals in an hour.

Q8. Generate a random sample of size 1000 from a Poisson distribution with mean 5 and calculate the sample mean and variance.

```python
import numpy as np

sample_size = 1000
mean = 5

random_sample = np.random.poisson(mean, sample_size)

sample_mean = np.mean(random_sample)
sample_variance = np.var(random_sample, ddof=0)  
print(f"Sample Mean: {sample_mean:.4f}")
print(f"Sample Variance: {sample_variance:.4f}")
```

Q9. How mean and variance are related in Binomial distribution and Poisson distribution?

Ans. Mean and variance are related in the Binomial and Poisson distributions in the following ways:

1. Binomial Distribution:

In a Binomial distribution, where you have a fixed number of trials (n) and a constant probability of success (p) in each trial, the mean (μ) and variance (σ²) are related as follows:

Mean (μ): The mean of a Binomial distribution is given by μ = n * p, where 'n' is the number of trials and 'p' is the probability of success in each trial.

Variance (σ²): The variance of a Binomial distribution is given by σ² = n * p * (1 - p), where 'n' is the number of trials, and 'p' is the probability of success in each trial.

In a Binomial distribution, as the number of trials 'n' increases while keeping the probability of success 'p' constant, both the mean and variance increase proportionally. The variance also depends on the probability of success and its complement (1 - p).

2. Poisson Distribution:

In a Poisson distribution, which is used to model rare and random events occurring in a fixed interval, the mean (μ) and variance (σ²) are related as follows:

Mean (μ): The mean of a Poisson distribution is denoted by μ, and it represents the average rate of event occurrences in the fixed interval.

Variance (σ²): The variance of a Poisson distribution is also equal to μ, which means σ² = μ.

In a Poisson distribution, the mean and variance are equal. This is a distinctive property of the Poisson distribution and is a result of the assumptions underlying the distribution, particularly the assumption that events occur independently and at a constant rate. If the rate of events (μ) is known, it completely determines both the mean and the variance of the distribution.

Q10. In normal distribution with respect to mean position, where does the least frequent data appear?

Ans. In a normal distribution, the least frequent data appears in the tails of the distribution, farthest from the mean. The normal distribution, also known as the Gaussian distribution, is symmetric and bell-shaped. The peak of the curve represents the mean of the distribution, and as you move away from the mean in either direction, the frequency of data decreases.

Specifically:

1. Data in the central region around the mean is the most frequent. This is where the majority of data points are concentrated, and it corresponds to values close to the mean.

2. Data in the tails of the distribution, which are farther from the mean, is less frequent. The tails of the normal distribution extend infinitely in both directions, and data points in the extreme tails are the least frequent.

3. The point at which the tails become extremely rare is quantified by the standard deviations from the mean. In a standard normal distribution (mean = 0, standard deviation = 1), the tails are considered to be the regions beyond approximately ±3 standard deviations from the mean. Beyond this point, the probability of observing data becomes very low.
