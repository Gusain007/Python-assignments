Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
Note: Also add title, xlabel, and ylabel to the plot.

```python
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
plt.title("This is a scatter graph")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.scatter(x,y)
```

Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
Use the following data:
import numpy as np
For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])
For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])
For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])
For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])

Ans. The subplot() function in Matplotlib is used to create multiple subplots within the same figure. It allows you to divide the figure into a grid of rows and columns and specify the position of each subplot within that grid. This is particularly useful when you want to display multiple plots side by side or in a grid layout.

```python
import matplotlib.pyplot as plt
import numpy as np


x = np.array([0, 1, 2, 3, 4, 5])

plt.figure(figsize=(12, 8))

# Subplot 1
plt.subplot(2, 2, 1)
y1 = np.array([0, 100, 200, 300, 400, 500])
plt.plot(x, y1)
plt.title('Line Plot 1')

# Subplot 2
plt.subplot(2, 2, 2)
y2 = np.array([50, 20, 40, 20, 60, 70])
plt.plot(x, y2)
plt.title('Line Plot 2')

# Subplot 3
plt.subplot(2, 2, 3)
y3 = np.array([10, 20, 30, 40, 50, 60])
plt.plot(x, y3)
plt.title('Line Plot 3')

# Subplot 4
plt.subplot(2, 2, 4)
y4 = np.array([200, 350, 250, 550, 450, 150])
plt.plot(x, y4)
plt.title('Line Plot 4')

plt.tight_layout()

plt.show()
```

Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

Ans. A bar plot, also known as a bar chart or bar graph, is a graphical representation of data where rectangular bars are used to represent categorical data points. The length or height of each bar is proportional to the value it represents. They are particularly useful for visualizing categorical data or showing comparisons between discrete items.

```python
import matplotlib.pyplot as plt
import numpy as np

company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

# Vertical Bar Plot
plt.figure(figsize=(5, 3))
plt.bar(company, profit, color='blue')
plt.xlabel('Company')
plt.ylabel('Profit (in millions)')
plt.title('Company Profits (Vertical Bar Plot)')
plt.show()

# Horizontal Bar Plot
plt.figure(figsize=(5, 3))
plt.barh(company, profit, color='green')
plt.xlabel('Profit (in millions)')
plt.ylabel('Company')
plt.title('Company Profits (Horizontal Bar Plot)')
plt.show()
```

Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

Ans. A box plot, also known as a box-and-whisker plot, is a graphical representation of the distribution of a dataset. It displays the summary statistics of a dataset in a concise and visual manner. Box plots are useful for visualizing the spread, central tendency, and skewness of data, as well as identifying any data points that may be considered outliers.

```python
import matplotlib.pyplot as plt
import numpy as np

box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

plt.figure(figsize=(5, 3))
plt.title('Box Plot of box1 and box2')
plt.xlabel('Box')
plt.ylabel('Value')

plt.boxplot([box1, box2], labels=['box1', 'box2'])

plt.show()
```
